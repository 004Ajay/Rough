```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
(X_train, y_train) , (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
print(len(X_train))
print(len(X_test))
```

```python
X_train[0].shape
```

```python
X_train[0]
```

```python
num = 0
plt.matshow(X_train[num])
plt.title(f"True Label: {y_train[num]}")
plt.axis('off')
plt.show()
```

```python
X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train[10]
```

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```python
X_train_flattened.shape
```

```python
X_train_flattened[0]
```

## Model with a single layer

```python
# Making model with a single layer (Dense Layer)

model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
acc = model.evaluate(X_test_flattened, y_test)
f"Accuray: {acc[1] * 100}" 
```

```python
y_predicted = model.predict(X_test_flattened)
plt.matshow(X_test[0])
f"Classification: {np.argmax(y_predicted[0])}"
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_predicted_labels)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Reds)
fig = disp.ax_.get_figure() 
fig.set_figwidth(6)
fig.set_figheight(6)
plt.show()
```

## New model with hidden layers

```python
# Making model with a hidden layers

model1 = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model1.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model1.fit(X_train_flattened, y_train, epochs=5)
```

```python
acc1 = model1.evaluate(X_test_flattened, y_test)
f"Accuray: {acc1[1] * 100}" 
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm1 = confusion_matrix(y_test, y_predicted_labels)

disp = ConfusionMatrixDisplay(confusion_matrix=cm1)
disp.plot(cmap=plt.cm.Greens)
fig = disp.ax_.get_figure() 
fig.set_figwidth(6)
fig.set_figheight(6)
plt.show()
```
