```python
def link_adder(df, url_mid_part):
    chemical = df[df.columns[0]]
    url_first_part = "https://cb.imsc.res.in/imppat/"    
    df['Link'] = chemical.apply(lambda x: f"{url_first_part}{url_mid_part}{x.replace(' ', '%20')}")
    return df
```

```python
df1 = pd.read_csv("IMPPAT_PhytoAssoc_Indian medicinal plants.csv")
df2 = pd.read_csv("IMPPAT_PhytoAssoc_Phytocemical.csv")
df3 = pd.read_csv("IMPPAT_PhytoAssoc_Chemical_Superclass.csv")
df4 = pd.read_csv("IMPPAT_Therapeutic_Indian medicinal plants.csv")
df5 = pd.read_csv("IMPPAT_Therapeutic_diseases.csv")

dfs = [df1, df2, df3, df4, df5]

file_names = [
    "IMPPAT_PhytoAssoc_Indian medicinal plants.csv",
    "IMPPAT_PhytoAssoc_Phytocemical.csv",
    "IMPPAT_PhytoAssoc_Chemical_Superclass.csv",
    "IMPPAT_Therapeutic_Indian medicinal plants.csv",
    "IMPPAT_Therapeutic_diseases.csv"
]

mid_urls = [
    'phytochemical/', 
    'phytochemicalplants/', 
    'phytochemical-searchsupclass/', 
    'therapeutics/', 
    'therapeuticsplants/'
]

for df, url_mid, file_name in zip(dfs, mid_urls, file_names):
    updated_df = link_adder(df, url_mid)
    
    updated_df.to_csv(file_name, index=False)
    print(f"Updated and saved: {file_name}")
```

```python
# Code with concurreny to process requests in multiple(50) threads

import pandas as pd
import concurrent.futures
import time
import random

urls = list(imppat_links_df['Link'])

full_df = pd.DataFrame()

def scrape_url(url):
    try:
        df = pd.read_html(url)[0]
        df['Phytochemical IMPPAT Link'] = 'https://cb.imsc.res.in/imppat/phytochemical-detailedpage/' + df['IMPPAT Phytochemical identifier']
        return df
        
    except Exception as e:
        print(f"Failed to scrape {url}: {e}")
        return None

def scrape_and_save(urls):
    global full_df
    
    with concurrent.futures.ThreadPoolExecutor(max_workers=50) as executor:    
        results = executor.map(scrape_url, urls) # Map the URLs to the scrape_url function

        for result in results:
            if result is not None:
                full_df = pd.concat([full_df, result], ignore_index=True)

            time.sleep(random.uniform(0.5, 2)) # a random delay (pause) for server
scrape_and_save(urls)

full_df.to_csv('IMPPAT_Indian_Plant_Tabular_Info.csv', index=False)
```
