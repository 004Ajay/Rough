```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import bokeh as bk
```

```python
# data
x = np.arange(0, 10)
y = np.arange(11, 21)
```

```python
# plotting using matplotlib
plt.scatter(x, y, c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Plotting using Matplotlib')
plt.show()
```

```python
# plotting using matplotlib
grf = sns.stripplot(x, y)
grf.set(xlabel='X axis', ylabel='Y axis', title='Plotting using Seaborn')
plt.show()
```
